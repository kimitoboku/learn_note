## 確率の入門
### 確率変数
確率変数はある不確定な量を表す。
その施行の結果や、温度の測定値、確率変数は施行を行うたびに異なる結果が得られる。
少しずつ値が変化するが、ある値の付近に集まるという性質がある。
これを確率分布で表すことができる。
確率関数の積は確率であるので1である。

確率密度関数は、取りうる確率をその面積で表したもの。

### 同時確率
２つ以上の事象が同時に発生する確率。
$Pr(x,y)$ や $P(x,y)$ と書かれる。
分布は等高線表示やヒートマップ表示などでプロットすることが出来る。

## 周辺化
xとyの同時分布が与えられた場合、xだけの分布、yだけの分布を求めることが出来る。

$$ Pr(x) = \int Pr(x,y) dy$$

$$ Pr(y) = \int Pr(x,y) dx$$

xの分布が欲しい場合はy方向を積分し、yの分布が欲しい場合はx方向で積分を行う。
離散の場合も、xとyの同時分布が与えられた場合に、xだけの分布が欲しい場合は、y方向の総和を求める。

xが離散でyが連続やその逆であった場合でも、積分や総和を求めることで、それぞれの分布を求めることが出来る。

確率変数が更に増えた場合であっても、離散であれば総和、連続であれば積分を行うことで、求めたい確率変数の分布を求めることが出来る。

### Question
#### Question1
以下の周辺化で得られる確率分布はどれか？

$$ \sum_{u} \int_{v} \sum_{t} \int_{z} Pr(s,t,u,v,w,x,y,z) dv dz $$

$Pr(s,w,x,y)$
